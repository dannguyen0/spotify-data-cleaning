```python
# Importing necessary packages for EDA with Python
import pandas as pd
```

```python
# Importing the 'spotify_tracks_data_2023' data set to start exploration
alldata = 'C:/Users/Dan/Documents/Professional/Python/Spotify/all_tracks_data.csv'
all_tracks = pd.read_csv(alldata)
pd.set_option('display.max_columns', None)
all_tracks.head(10)
```

```python
# Removing Unwanted Columns
all_tracks = all_tracks.drop(columns = ['artists', 'analysis_url', 'key', 'mode', 'time_signature', 'track_href', 'type', 'uri', 'rn'])
all_tracks.head(10)
```

```python
# Updating data types for specific columns
all_tracks['release_date'] = pd.to_datetime(all_tracks['release_date'])
all_tracks['release_date'] = all_tracks['release_date'].dt.tz_localize(None)
all_tracks.info()
```

```python
all_tracks.head(10)
```

```python
# Finding total counts of tracks that are explicit vs. non-explicit
all_tracks.value_counts('explicit')
```

```python
# Updating data type for explicit column to boolean
all_tracks['explicit'] = all_tracks['explicit'].convert_dtypes(infer_objects =True)
all_tracks.head()
```

```python
# Finding value counts for tracks based on popularity, very few are popular amongst listeners
all_tracks.value_counts('track_popularity')
```

```python
# Updating release year to being a date data type
all_tracks['release_year'] = pd.to_datetime(all_tracks['release_year'], format = '%Y')
all_tracks['release_year'] = all_tracks['release_year'].dt.to_period('y')
```

```python
all_tracks.info()
```

```python
# Finding the most popular tracks based on track popularity column 
all_tracks.sort_values(by = 'track_popularity', ascending = False).head(10)
```

```python
# Renaming artists and genre columns to improve user friendliness
artistmap = {"artist_0" : "artist_1", 
             "artist_1" : "artist_2", 
             "artist_2" : "artist_3", 
             "artist_3" : "artist_4",
             "artist_4" : "artist_5"}

genremap = {"genre_0" : "genre_1",
            "genre_1" : "genre_2",
            "genre_2" : "genre_3",
            "genre_3" : "genre_4",
            "genre_4" : "genre_5"}

all_tracks.rename(artistmap, axis = 1, inplace = True)
all_tracks.rename(genremap, axis = 1, inplace = True)
all_tracks.head()
```

```python
# Adding new columns that counts the total non-null values from the artists and genre columns
all_tracks['artist_count'] = all_tracks.loc[:, 'artist_1':'artist_5'].count(axis = 1)
all_tracks['genre_count'] = all_tracks.loc[:, 'genre_1':'genre_5'].count(axis = 1)
all_tracks.head()
```

```python
# Exporting the final changes to a new Excel workbook to import into Power BI
all_tracks.to_excel('spotify_tracks_final.xlsx', sheet_name = 'Tracks')
```
